```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
df=pd.read_csv('/Users/saurabh/Downloads/Airline_Delay_Cause_20250604_001439/Airline_Delay_Cause.csv',index_col='carrier_name')
df=df.drop(columns=['year','month'])
df
```

```python
total_flights_by_airline = df.groupby('carrier_name')['arr_flights'].sum().astype(int).sort_values(ascending=False)
print('The Total flights by each arline in descending order are:\n',total_flights_by_airline)
print()
total_delay_by_airline=df.groupby('carrier_name')['arr_del15'].sum().astype(int).sort_values(ascending=False)
print('The Total delayed flights by each arline in descending order are:\n',total_delay_by_airline)
```

```python
delay_rate_by_airline=(total_delay_by_airline/total_flights_by_airline)*100
delay_rate_by_airline = delay_rate_by_airline.round(2)
print('The percent delay rate by Airline is:\n',delay_rate_by_airline)
delay_rate_by_airline.sort_values(ascending=True).plot(kind='barh', color='red',title='Airline Delay Rate Comparison',xlabel='Delay Rate',ylabel='Airline')
```

```python
total_flights_by_airport=df.groupby('airport_name')['arr_flights'].sum().astype(int).sort_values(ascending=False)
print('The Total flights by each airport in descending order are:\n',total_flights_by_airport)
print()
total_delay_by_airport=df.groupby('airport_name')['arr_del15'].sum().astype(int).sort_values(ascending=False)
print('The Total delayed flights by each airport in descending order are:\n',total_delay_by_airport)
```

```python
delay_rate_by_airport=(total_delay_by_airport/total_flights_by_airport)*100
delay_rate_by_airport=delay_rate_by_airport.round(2)
print('The percent delay rate by Airport is:\n',delay_rate_by_airport)
delay_rate_by_airport.sort_values(ascending=True).head(20).plot(kind='barh', color='red',title='Top 20 Airports with highest Delay Rates',xlabel='Delay Rate',ylabel='Airport')
#make a treemap
```

```python
delay_rate_by_airport.sort_values(ascending=False).head(20).plot(kind='barh', color='green',title='Top 20 Airports with lowest Delay Rates',xlabel='Delay Rate',ylabel='Airport')
```

```python
delay_causes = ['carrier_ct', 'weather_ct', 'nas_ct', 'security_ct', 'late_aircraft_ct']
total_by_cause = df[delay_causes].sum().sort_values(ascending=False)
```

```python
# Mapping from column names to readable labels
label_map = {
    'carrier_ct': 'Carrier',
    'weather_ct': 'Weather',
    'nas_ct': 'NAS (Airspace)',
    'security_ct': 'Security',
    'late_aircraft_ct': 'Late Aircraft'
}
# Rename the index for readability
total_by_cause.index = total_by_cause.index.map(label_map)
# Plot with readable labels
total_by_cause.plot(kind='pie',autopct='%1.1f%%',figsize=(8, 8),title='Causes of Delay Distribution')
```
